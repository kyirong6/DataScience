```python
import pandas as pd
from scipy import stats as sp
import matplotlib.pyplot as plt
```

```python
tweets = pd.read_csv("dog_rates_tweets.csv", sep=',', parse_dates=['created_at'])
```

```python
regex = r'(\d+(\.\d+)?)/10'
tweets = tweets[tweets.text.str.contains(regex)]
tweets['rating'] = pd.to_numeric(tweets.text.str.extract(regex)[0], downcast='float')
tweets = tweets[tweets.rating <= 25]
tweets = tweets[tweets.rating >= 0]
```

```python
def to_timestamp(x):
    return x.timestamp()

tweets['timestamp'] = tweets['created_at'].apply(to_timestamp)
```

```python
lr = sp.linregress(tweets.timestamp, tweets.rating)
tweets['prediction'] = (tweets.timestamp * lr.slope) + lr.intercept
```

```python
plt.figure(figsize=(10, 7))
plt.xticks(rotation=25)
plt.xlabel("Date")
plt.ylabel("Rating")
plt.scatter(tweets.created_at, tweets.rating, color='b', label="original data", alpha=.5)
plt.plot(tweets.created_at, lr.intercept + lr.slope*tweets.timestamp, color='r', label='fitted line', linewidth=3)
plt.legend()
plt.show()
```
